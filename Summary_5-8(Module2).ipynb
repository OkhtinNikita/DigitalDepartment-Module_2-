```python
# графики можно выводить с различными параметрами (например разными цветами)
# Цвета линий color, можно задавать:
# - с помощью констант, 'blue'
# - 'rbgcmyk' - с помощью rgb
# - градации серого от 0 до 1
# - RRGGBB - 'FF00EE'ArithmeticError
# - RGB - (1.0, 0.2, 0.3)
# - HTML - 'salmon

# Стиль линии, line
# - сплошная '-', 'solid'
# - 'dotted'
# - 'dashed'
# - 'dashdot'

fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x), color = 'blue', linestyle = 'solid')
ax.plot(x, np.sin(x-1), color = 'g')
ax.plot(x, np.sin(x-2), color = '#FF00EE', linestyle = 'dashed')
ax.plot(x, np.sin(x-3), linestyle = 'dotted')
ax.plot(x, np.sin(x-4), linestyle = 'dashdot')
ax.plot(x, np.sin(x-5))

plt.show()
```

```python
## Поясняющие надписи

births = pd.read_csv("./data/births-1969.csv")

births["day"] = births["day"].astype(int)  # Преобразует дни в тип int
births.index = pd.to_datetime(
    10000 * births.year + 100 * births.month + births.day, format="%Y%m%d"
)
# print(births.head())

births_by_date = births.pivot_table("births", [births.index.month, births.index.day])
# print(births_by_date.head())
from datetime import datetime

births_by_date.index = [
    datetime(1969, month, day) for (month, day) in births_by_date.index
]
print(births_by_date.head())

fig, ax = plt.subplots()
births_by_date.plot(ax=ax)

style = dict(size=10, color="gray")
ax.text("1969-01-01", 5500, "Новый Год!", **style)
ax.text("1969-09-01", 4500, "День Знаний!", ha="right")

ax.set(title="Рождаемость в 1969", ylabel="Число рождений")

ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'))
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))


plt.show()
```

```python
# Поверхностный
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


fig = plt.figure()
ax = plt.axes(projection="3d")
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.plot_surface(X, Y, Z, cmap='viridis',edgecolor='None')
ax.set_title('Example')
```
